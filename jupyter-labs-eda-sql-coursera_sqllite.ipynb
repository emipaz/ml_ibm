### Tarea 2

##### Mostrar 5 registros en los que los sitios de lanzamiento comienzan con la cadena 'CCA'

### Tarea 3

##### Mostrar la masa total de la carga útil transportada por los boosters lanzados por la NASA (CRS)


```python
%sql select sum("PAYLOAD_MASS__KG_") as suma from SPACEXTBL where customer like 'NASA (CRS)'
```

### Tarea 4

##### Visualizar la masa media de la carga útil transportada por la versión del booster F9 v1.1


```python
%%sql 
select avg(payload_mass__kg_) as Promedio 
from SPACEXTBL 
where booster_version like 'F9 v1.1%'
```

### Tarea 5

##### Enumere la fecha en la que se logró el primer resultado exitoso de aterrizaje en la plataforma de tierra.

*Sugerencia: Utilice la función min*.


```python
minimo = %sql select min(date) as Date from SPACEXTBL where mission_outcome like 'Success'
type(minimo.DataFrame())
minimo
```

### Tarea 6

##### Enumere los nombres de los propulsores que tienen éxito en la nave no tripulada y tienen una masa de carga útil superior a 4000 pero inferior a 6000


```python
%%sql 
select booster_version 
from SPACEXTBL 
WHERE (mission_outcome like 'Success') AND 
(payload_mass__kg_ BETWEEN 4000 AND 6000) 
AND ("Landing _Outcome" like 'Success (drone ship)')
```

### Task 7

##### List the total number of successful and failure mission outcomes


```python
%%sql 
SELECT mission_outcome, count(*) as Count 
FROM SPACEXTBL 
GROUP by mission_outcome 
ORDER BY mission_outcome
```

### Tarea 8

##### Enumere los nombres de las versiones de booster que han transportado la masa máxima de carga útil. Utilice una subconsulta


```python
maxm = %sql select max(payload_mass__kg_) from SPACEXTBL
print(maxm)
maxv = maxm[0][0]
print(maxv)
%sql select booster_version from SPACEXTBL where payload_mass__kg_= (select max(payload_mass__kg_) from SPACEXTBL)
```

### Tarea 9

##### Listar los registros que mostrarán los nombres de los meses, los resultados de aterrizaje de los fallos en la nave de drones, las versiones de los propulsores y el lugar de lanzamiento para los meses del año 2015.

**Nota: SQLLite no soporta nombres de meses. Por lo tanto, es necesario utilizar substr(Date, 4, 2) como mes para obtener los meses y substr(Date,7,4)='2015' para el año.**


```python
%%sql
select substr(DATE, 6, 2) as Mes ,substr(Date,0,5) as Año, "Landing _Outcome" , booster_version, launch_site 
from SPACEXTBL
where substr(Date,0,5) like '2015%' AND 
"Landing _Outcome" like 'Failure (drone ship)' 
```

### Tarea 10

##### Clasifique el recuento de landing_outcomes exitosos entre la fecha 04-06-2010 y el 20-03-2017 en orden descendente.


```python
%%sql 
SELECT "Landing _Outcome", count(*) as Count 
FROM SPACEXTBL
where substr(Date,0,5) >= '2010-06-04' AND substr(Date,0,5) <= '2017-03-20'
GROUP by "Landing _Outcome" ORDER BY count Desc
```
